##Exercise: Finding no. of points outside nd sphere

```python
nd = 2
npoints = 100000
```

```python
points = np.random.rand(npoints, nd)
points
```

```python
#distance from origin
dfo = np.zeros((npoints, 1))
dfo
outside_points = 0
```

```python
%%time
for i in range(npoints):
    for j in range(nd):
        dfo[i][0] += points[i, j] ** 2
    dfo[i][0] = np.sqrt(dfo[i])
    if dfo[i][0] > 1:
        outside_points += 1
```

```python
# 1 - (3.14/4) = 0.214999997
print('Fraction of points outside is: ', outside_points/npoints )
```

```python
dfo[ dfo > 1]
```

```python
%%time
sq_points = points * points
dfo = np.sqrt(np.sum(sq_points, axis = 1)) # rows: axis - 0, columns: axis - 1
outside_points = np.sum(dfo > 1)
```

```python
print('Fraction of points outside is: ', outside_points/npoints )
```

```python
%%time
outside_points = np.sum( np.sqrt(np.sum(points * points, axis = 1)) > 1)
```

```python
def area_outside_square(npoints, nd):
    points = np.random.rand(npoints, nd)
    return np.sum( np.sqrt(np.sum(points * points, axis = 1)) > 1) / npoints
```

```python
area_outside_square(100000, 2)
```

```python
for i in range(2, 11):
    print(i, area_outside_square(100000, i))
```
